```python
## Part A

nums = [3, 7, 1, 7, 2]
print(nums[0], nums[4])

nums.append(10)
nums.insert(0,0)
print(nums)

nums[1] = 99

print(nums.count(7))

## Part B

student = ("A101", "Ana", 19)   # (id, name, age)

var1, var2, var3 = student
print(var1)
print(var2)
print(var3)

## student[2] = 20, tuple assignments can not be individually changed, only fully reassigned as a new tuple (immutable)

student2 = ("A102", "Boo", 18)
roster = [student2, student2]
print(roster)

## Part C

prices = {}

prices['apple'] = 1.25
prices['banana'] = 0.79
prices['milk'] = 2.99
print(prices)

print(prices['banana'])

prices['milk'] = 3.19
print(prices)

print(prices.keys())
print(prices.values())
```
