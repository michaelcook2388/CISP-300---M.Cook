```python
##List Comprehension

num = list(range(1, 10, 2))
print(num)

newList = [x * x for x in num] 
##Same as for loop
##for x in num:
##    x + 5 
print(newList)

new_list = [x + 5 for x in num]
print(new_list)
```

```python
num = list(range(10, 0, -1))
print(num)

newList = [x - 10 for x in num]
print(newList)
```

```python
num = list(range(1,100))
print(num)
print(len(num))

newList = [x for x in num if x % 2 == 0]
print(newList)
print(len(newList))
```

```python
## If-Else Statements

year = int(input())

if year < 1901:
    print('Long ago')
if year == 2025:
    print('This Year')
elif year < 2001:
    print('20th century')
elif year < 2101:
    print('21st century')
elif year < 2201:
    print('22nd century')
else:
    print('Distant Future')
```
